```python
import requests
import bs4
import pandas as pd
```

```python
# Collecting stock links from Groww
stock_link = []
base_url = 'https://groww.in'
for page_num in range(0,289):
    re = requests.get(f'https://groww.in/stocks/filter?closePriceHigh=100000&closePriceLow=0&marketCapHigh=2000000&marketCapLow=0&page={page_num}&size=15&sortType=ASC')
    soup = bs4.BeautifulSoup(re.text, "html.parser")
    for i in soup.find_all('td',class_='bodyBaseHeavy'):
        stock_link.append(base_url + i.find('a').get('href'))
```

```python
stock_link
```

```python
# Store stocks link in text file
with open(r'links4335.txt','w') as f:
    for link in stock_link:
        f.write(link+'\n')
```

```python
# Load stocks link from text file
urls =[]
with open(r'links4335.txt','r') as f:
    for link in f.readlines():
        urls.append(link.strip())
```

```python
len(urls)
```

```python
import json

industry = []
stock_name = []

market_cap = []
pb_ratio = []
pe_ratio = []
div_yield = []
book_value = []
eps = []
roe = []
cap = []

def check(lis,dic,key):
    if key in dic:
        lis.append(dic[key])
    else:
        lis.append(None)

def check2(lis,dic,key):
    if not dic:
        lis.append(dic)
    else:
        lis.append(dic.get(key))

for num,url in enumerate(urls):
    print(num,url)
    req = requests.get(url)
    s = bs4.BeautifulSoup(req.text,'html.parser')
    json_data = s.select('script#__NEXT_DATA__')[0].text
    json_format = json.loads(json_data)
    
    header = json_format['props']['pageProps']['stockData']['header']
    check(industry,header,'industryName')


    stock_name.append(header['displayName'])

    stats = json_format['props']['pageProps']['stockData'].get('stats')
    check2(market_cap,stats,'marketCap')
    check2(pb_ratio,stats,'pbRatio')
    check2(pe_ratio,stats,'peRatio')
    check2(div_yield,stats,'divYield')
    check2(book_value,stats,'bookValue')
    check2(eps,stats,'epsTtm')
    check2(roe,stats,'roe')
    check2(cap,stats,'cappedType')
```

```python
stocks_df = pd.DataFrame({'stock_name':stock_name,
                          'industry':industry, 'cap':cap, 'market_cap':market_cap, 'pe_ratio':pe_ratio,
                          'pb_ratio':pb_ratio, 'div_yield':div_yield,'book_value':book_value,'eps':eps,'roe':roe})
```
